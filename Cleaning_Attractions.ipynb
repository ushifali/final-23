```python
import pandas as pd
import numpy as np
     
# Bangalore data
attraction_df_1 = pd.read_csv('/content/drive/MyDrive/dATA/dataset_crawler-google-places_2023-03-21_14-01-20-553.csv',encoding='utf-8')

#Udupi Data
attraction_df_2 = pd.read_csv('/content/drive/MyDrive/dATA/dataset_crawler-google-places_2023-03-21_15-46-23-957.csv',encoding='utf-8')

#Mangalore Data
attraction_df_3 = pd.read_csv('/content/drive/MyDrive/dATA/dataset_crawler-google-places_2023-03-21_15-59-07-566.csv',encoding='utf-8')
```

```python
attraction_df = pd.concat([attraction_df_1, attraction_df_2, attraction_df_3])

# reset the index of the combined DataFrame
attraction_df = attraction_df.reset_index(drop=True)
attraction_df
```

```python
attraction_df = attraction_df.drop(attraction_df.filter(regex='updatesFromCustomers|^additionalInfo').columns, axis=1)
attraction_df
```

```python
# To concatenate values from the categories/0, categories/1, categories/2, categories/3, categories/4, categories/5,
#  and categories/6 columns into a single categories column while ignoring the NaN values,
#  you can use the apply() method in pandas along with a lambda function. Here's the code to do so: 
attraction_df['categories'] = attraction_df.apply(lambda x: ', '.join([str(x[f'categories/{i}']) \
                                                                       for i in range(7) if pd.notnull(x[f'categories/{i}'])]), axis=1)
```

```python
attraction_df.drop(['categories/0', 'categories/1', 'categories/2', 'categories/3', 'categories/4', 'categories/5', 'categories/6'], axis=1, inplace=True)
```

```python
print(attraction_df.columns)
```

```python
columns_to_drop = ['cid', 'claimThisBusiness','countryCode','imagesCount','temporarilyClosed','subTitle']
attraction_df.drop(columns_to_drop , axis=1, inplace=True)
print(attraction_df.columns)
```

```python
attraction_df = attraction_df.drop(attraction_df.filter(regex='^peopleAlsoSearch|^placesTags|^popularTimesHistogram|^searchPageLoadedUrl|^scrapedAt|^searchString').columns, axis=1)
attraction_df
```

```python
attraction_df = attraction_df.drop(attraction_df.filter(regex='^isAdvertisement|^openingHours|^menu|^orderBy|^permanentlyClosed|^questionsAndAnswers').columns, axis=1)
```

```python
columns_to_drop = ['neighborhood', 'placeId','plusCode','popularTimesLivePercent','popularTimesLiveText','postalCode','searchPageUrl']
attraction_df.drop(columns_to_drop , axis=1, inplace=True)
```

```python
columns_to_drop = ['website', 'url','locatedIn','categoryName','phone']
attraction_df.drop(columns_to_drop , axis=1, inplace=True)
```

```python
columns_to_drop = [	'rank',	'reviewsCount'	]
attraction_df.drop(columns_to_drop , axis=1, inplace=True)
```

```python
attraction_df = attraction_df.drop(attraction_df.filter(regex='^reviews/').columns, axis=1)
```

```python
attraction_df = attraction_df.drop(attraction_df.filter(regex='^reviewsDistribution|^reviewsTags').columns, axis=1)
```

```python
attraction_df = attraction_df.rename(columns={'location/lat': 'latitude', 'location/lng': 'longitude'})
```

```python
#To concatenate the values of the imageUrls/0 and imageUrls/1 columns into a single column named image_url, 
# you can use the apply() method along with a lambda function in pandas.
attraction_df['image_url'] = attraction_df[['imageUrls/0', 'imageUrls/1']].apply(lambda x: ','.join(x.dropna().astype(str)), axis=1)
```

```python
attraction_df = attraction_df.drop(['imageUrls/0', 'imageUrls/1'], axis=1)
```
