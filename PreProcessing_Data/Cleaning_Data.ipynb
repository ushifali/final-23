```python
df7= pd.read_csv('/content/drive/MyDrive/dATA/makemytrip_com-travel_sample.csv')
```

```python
df7.head()
```

```python
columns_to_drop = ['country','mmt_review_count','mmt_review_rating','mmt_review_score','mmt_traveller_type_review_count','crawl_date','is_value_plus','mmt_location_rating','mmt_tripadvisor_count','query_time_stamp','qts','site_review_count','site_review_rating','property_id','sitename','uniq_id','mmt_holidayiq_review_count']


# drop the columns you want to remove
df7 = df7.drop(columns_to_drop, axis=1)

```

```python
import numpy as np
```

```python
df7 = df7.rename(columns={'highlight_value': 'hotel_facilities'})
df7 = df7.rename(columns={'property_address': 'address'})
df7 = df7.rename(columns={'hotel_overview': 'hotel_description'})
df7 = df7.rename(columns={'traveller_rating': 'rating'})
df7 = df7.rename(columns={'room_types': 'room_type'})
df7 = df7.replace('{{value}}', np.nan)
df7 = df7.replace('{{facility}}', np.nan)

df7['hotel_facilities']=df7['hotel_facilities']+"|"+ df7['in_your_room']
df7 = df7.drop('in_your_room', axis=1)
```

```python
df7.head(50)
```

```python
df7['rating'] = df7['rating'].str.split(':').str[1].str.split('/').str[0].astype(float)
```

```python
df7.head(50)
```

```python
df7.to_csv('/content/drive/MyDrive/Cleaned_Data/MakeMyTrip.csv', index=False)
```
