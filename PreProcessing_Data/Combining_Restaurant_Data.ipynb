```python
# read the first CSV file
df1 = pd.read_csv('/content/drive/MyDrive/final_year_project/data/Cleaned_Data/Bangalore_Restaurants.csv')
```

```python
df1.head()
```

```python
df1 = df1.drop(["Dining_Review_Count"], axis=1)
```

```python
df1.head()
```

```python

# read the second CSV file
df4 = pd.read_csv('/content/drive/MyDrive/final_year_project/data/Cleaned_Data/zomato_restaurants_in_India.csv')
```

```python
df4.head()
```

```python
df4['establishment']=df4['establishment'].str[2:-2]
df4 = df4.drop(['zipcode'], axis=1)
df4.head()
```

```python
df4['cuisines']=df4['cuisines']+" - "+df4['establishment']
df4 = df4.drop(['city','locality','rating_text','establishment'], axis=1)
```

```python
df4.head()
```

```python
df4=df4.rename(columns={'name': 'Restaurant_Name', 'cuisines': 'Cuisines','url':'Website','address':'Address','locality_verbose':'Locality','latitude':'Latitude','longitude':'Longitude','average_cost_for_two':'Pricing_for_2','aggregate_rating':'Dining_Rating','timings':'Timings','highlights':'Highlights','votes':'Votes','delivery':'Delivery','takeaway':'Takeaway'})
df4.head()
```

```python
df1=df1.rename(columns={'Category':'Cuisines'})
```

```python
df1.head()
```

```python
df5=pd.concat([df1,df4])
df5.head()
```

```python
df5=df5.drop_duplicates()
df5.head()
```

```python
list_df5=df5.columns.to_list()
list_df5.remove('Restaurant_Name')
# print(list_df5)
aggr_func= lambda x: x.dropna().iloc[0] if len(x.dropna()) > 0 else np.nan
df5_groupby_arg_dict= { value: aggr_func for value in list_df5}
df5_groupby_arg_dict['Dining_Rating']=['mean']
df5_groupby_arg_dict['Cuisines']=lambda x: " ,".join(str(i) for i in x)
df5_groupby_arg_dict['Pricing_for_2']= ['mean']
print(df5_groupby_arg_dict)
```

```python
aggr_func= lambda x: x.head(1) if x not in ['Dining_rating','Cuisines','Pricing_for_2']
```

```python

df5_sample=df5.groupby(['Restaurant_Name']).agg(df5_groupby_arg_dict)
df5_sample.head(50)
```

```python
df5_data = df5_sample.copy()  
df5_data.reset_index(drop=False, inplace=True)  # reset the index to a column of sequential numbers 
```

```python
df5_data.head()
```

```python
df5_data.to_csv('/content/drive/MyDrive/final_year_project/data/Cleaned_Data/combined/combined_restaurants.csv', index=False)
```
