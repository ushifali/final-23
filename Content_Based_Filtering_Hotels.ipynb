```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cdist
```

```python
# Load data
data = pd.read_csv('/content/drive/MyDrive/Cleaned_Data/combined/final_hotel_dataset.csv')
data.head(20)
```

```python
# Define user location
user_location = np.array([12.970394, 77.664713])  # Bangalore

# Calculate distance between user location and all places
places_location = data[['latitude', 'longitude']].values
distances = cdist(user_location.reshape(1, -1), places_location, 'euclidean')[0]

# Sort places by distance and recommend top 5
recommended_places = data.iloc[distances.argsort()[:50]]
recommended_places.head(10)
```

```python
data['Distance'] = distances
data.head()
```

```python
#taking rating from the hotel star rating 
# Extract the number using a regular expression
data['hotel_star_rating'] = data['hotel_star_rating'].str.extract(r'(\d+)').astype(float)
data['hotel_star_rating'] = data['hotel_star_rating'].fillna(0).astype(int)
```

```python
import pandas as pd
from annoy import AnnoyIndex
from sklearn.preprocessing import StandardScaler

attractions_df = data.copy()

# Select the relevant columns and drop missing values
selected_cols = ['Distance', 'rating','hotel_star_rating']
attractions_df = attractions_df[selected_cols].dropna()

# Convert string features to numerical values
for column in selected_cols:
    attractions_df[column] = pd.to_numeric(attractions_df[column], errors='coerce').fillna(0)


# Normalize the data using StandardScaler
scaler = StandardScaler()
attractions_df[selected_cols] = scaler.fit_transform(attractions_df[selected_cols])

# # Scale the 'distance' feature
# from sklearn.preprocessing import MinMaxScaler
# scaler1 = MinMaxScaler()
# attractions_df['Distance'] = scaler1.fit_transform(attractions_df[['Distance']])

# Create an AnnoyIndex for the normalized data
n_trees = 100
f = len(selected_cols)
annoy_index = AnnoyIndex(f, metric='euclidean')
for i, row in attractions_df.iterrows():
    features = row[selected_cols].values.astype('float32')
    annoy_index.add_item(i, features)
annoy_index.build(n_trees)


```

```python
def get_recommendations(user_prefs, max_distance=10, k=50):
    user_distance = user_prefs['Distance']
    filtered_attractions = attractions_df[attractions_df['Distance'] <= max_distance]
    filtered_indices = filtered_attractions.index

    user_prefs = user_prefs[selected_cols].values.astype('float32')
    user_prefs = scaler.transform(user_prefs.reshape(1, -1))

    # Query the AnnoyIndex for the most similar attractions
    nearest_indices = annoy_index.get_nns_by_vector(user_prefs.reshape(-1), k+len(filtered_indices))

    # Filter the recommendations based on the distance feature
    nearest_indices = [i for i in nearest_indices if i in filtered_indices][:k]

    # Return the recommended attractions
    return data.iloc[nearest_indices]
  
# Test the function with some example user preferences
user_prefs = pd.Series({'Distance': 0,'rating':4.5, 'hotel_star_rating':2})
recommended_attractions = get_recommendations(user_prefs)
# print(f"Recommended attractions: {recommended_attractions}")
recommended_attractions
```
